## Statistik

```python
import numpy as np
import seaborn as sns
import statistics as stats
import matplotlib.pyplot as plt
import math
```

## Measure of Central Value
Mean, Median, Modus

```python
data1 = [1,1,1,2,3,4,5,6]
n = len(data1)

mean = sum(data1)/n
median = stats.median(data1)
modus = stats.mode(data1)

print(f"Data: {data1}")
print(f"Banyak Data: {n}")
print(f"Mean Data: {mean}")
print(f"Median Data: {median}")
print(f"Modus Data: {modus}")
```

## Measure of Spread 1
Range, Quartile, Interquartile Range

```python
data2 = [3, 4, 4, 4, 7, 10, 11, 12, 14, 16, 17, 18, 35] # Pastikan data sudah diurut dari yang terkecil hingga terbesar
n = len(data2)

max = np.max(data2)
min = np.min(data2)
ranges = max - min

q1 = np.percentile(data2, 25) # Median data bawah
q2 = np.percentile(data2, 50) # Median data         
q3 = np.percentile(data2, 75) # Median data atas

iqr = q3 - q1

lowerBound = q1 - 1.5 * iqr # Batas bawah
upperBound = q3 + 1.5 * iqr # Batas atas

# Boxplot
plt.figure(figsize=(6, 4))
plt.title("Box Plot dengan Outliers")
plt.xlabel("Values")
plt.boxplot(data2, vert=False)

print(f"Max: {max}, Min: {min}, Range: {ranges}")
print(f"Q1: {q1}, Q2: {q2}, Q3: {q3}")
print(f"IQR: {iqr}")
print(f"Batas Bawah: {lowerBound}")
print(f"Batas Atas: {upperBound}")
```

## Measure of Spread 2
Varians, Standard Deviasi, Standard Error

```python
def variance(data, mean):
    n = len(data)
    total = 0
    for i in range(n):
        total += pow((data[i]-mean),2)
    return total/n

data3 = [600,470,170,430,300]
n = len(data3)

mean_3 = sum(data3)/n
varians = variance(data3, mean_3)

std = math.sqrt(varians)
se = std/math.sqrt(n)

print(f"Data: {data3}")
print(f"Banyak Data: {n}")
print(f"Mean Data: {mean_3:1.0f}")
print(f"Varians: {varians:1.0f}")
print(f"Standard Deviasi: {std:0.2f}")
print(f"Standard Error: {se:0.2f}")
```

## Measure of Shape
Skewness dan Kurtosis

```python
def skewness(data, mean, std):
    n = len(data)
    total = 0
    for i in range(n):
        total += pow((data[i]-mean),3)

    return total/(n * pow(std,3))

def kurtosis(data, mean, std):
    n = len(data)
    total = 0
    for i in range(n):
        total += pow((data[i]-mean),4)

    return total/(n * pow(std,4))

data_x = [20,10,10,45,25,60,55,18,70,15,90]
n = len(data_x)

mean_x = sum(data_x)/n
median_x = stats.median(data_x)
modus_x = stats.mode(data_x)

varians_x = variance(data_x, mean_x) # Menggunakan fungsi Variance sebelumnya

std_x = math.sqrt(varians_x)

skew_x = skewness(data_x, mean_x, std_x)

kurto_x = kurtosis(data_x, mean_x, std_x)

# Plot
sns.kdeplot(data_x, color="black")
sns.despine(top=True, right=True, left=True)
plt.yticks([])
plt.ylabel("")
plt.xlabel("")
plt.title("Data X", fontdict=dict(fontsize=10))
plt.axvline(mean_x, color="red", label="Mean")
plt.axvline(median_x, color="green", label="Median")
plt.axvline(modus_x, color="blue", label="Mode")
plt.legend()

print(f"Data: {data_x}")
print(f"Banyak Data: {n}")
print(f"Mean Data: {mean_x:0.2f}")
print(f"Median Data: {median_x:0.2f}")
print(f"Modus Data: {modus_x}")
print(f"Varians: {varians_x:0.2f}")
print(f"Standard Deviasi: {std_x:0.2f}")
print(f"Skewness: {skew_x:0.2f}")
print(f"Kurtosis: {kurto_x:0.2f}")
```
