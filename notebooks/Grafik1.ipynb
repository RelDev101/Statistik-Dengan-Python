## Grafik dengan Python
Bar, Pie, Line, Histogram, Scatter Plot.

```python
import matplotlib.pyplot as plt
```

```python
Grade = ["A", "B", "C", "D"]
Nilai = [11,15,4,6]

# Bar Chart
plt.bar(x=Grade, height=Nilai)
plt.title(label="Bar Chart")
```

```python
# Pie Chart
plt.pie(labels=Grade, x=Nilai)
plt.title(label="Pie Chart")
print()
```

```python
X = ["2011","2012","2013","2014","2015","2016","2017","2018","2019","2020"]
Y = [100,250,300,200,350,400,300,450,500,400]

# Line Chart
plt.plot(X, Y)
plt.title("Line Chart")
```

```python
data = [23, 45, 56, 78, 34, 67, 89, 45, 23, 45, 67, 89, 90, 34, 56]

# Histogram
plt.hist(data, bins=5)
plt.title("Histogram")
```

```python
# Scatter Plot
plt.scatter(X, Y)
plt.title("Scatter Plot")
```
